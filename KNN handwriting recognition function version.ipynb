```python
import numpy as np
import operator
from os import listdir
from sklearn.neighbors import KNeighborsClassifier as kNN
```

```python
def img2vector(filename):
    returnVect = np.zeros((1, 1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0, 32*i+j] = int(lineStr[j])
    return returnVect
```

```python
def handwritingClassTest():
    hwLabels = []
    trainingFileList = listdir('C:/Users/Owner/Downloads/Machine-Learning-master/Machine-Learning-master/kNN\handwriting/trainingDigits')
    m = len(trainingFileList)
    trainingMat = np.zeros((m, 1024))
    for i in range(m):
        fileNameStr = trainingFileList[i]
        classNumber = int(fileNameStr.split('_')[0])
        hwLabels.append(classNumber)
        trainingMat[i,:] = img2vector('C:/Users/Owner/Downloads/Machine-Learning-master/Machine-Learning-master/kNN\handwriting/trainingDigits/%s' % (fileNameStr))
    neigh = kNN(n_neighbors = 3, algorithm = 'auto')
    neigh.fit(trainingMat, hwLabels)
    testFileList = listdir('C:/Users/Owner/Downloads/Machine-Learning-master/Machine-Learning-master/kNN\handwriting/testDigits')
    errorCount = 0.0
    mTest = len(testFileList)
    for i in range(mTest):
        fileNameStr = testFileList[i]
        classNumber = int(fileNameStr.split('_')[0])
        vectorUnderTest = img2vector('C:/Users/Owner/Downloads/Machine-Learning-master/Machine-Learning-master/kNN\handwriting/testDigits/%s' % (fileNameStr))
        classifierResult = neigh.predict(vectorUnderTest)
        print("Prediction:%d,Truth:%d" % (classifierResult, classNumber))
        if(classifierResult != classNumber):
            errorCount += 1.0
    print(" the total number of errors is:%d\nthe total error rate is:%f%%" % (errorCount, errorCount/mTest * 100))
```

```python
if __name__ == '__main__':
    handwritingClassTest()
```
